## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

![](https://github.com/lvzeyu/chss_2025_assignment/blob/main/Assignment8/Figure1.png?raw=1)

```python
age = df["age"]

#グラフの数を設定
fig, ax =plt.subplots(2,2,figsize=(12,6))

ax1 = ax[0,0]
ax2 = ax[0,1]
ax3 = ax[1,0]
ax4 = ax[1,1]

#グラフの形式をそろえる
##軸ラベル
ax1.set_xlabel('年齢(歳)',fontsize=12)
ax1.set_ylabel('人数',fontsize=12)
ax2.set_xlabel('年齢(歳)',fontsize=12)
ax2.set_ylabel('人数',fontsize=12)
ax3.set_xlabel('年齢(歳)',fontsize=12)
ax3.set_ylabel('人数',fontsize=12)
ax4.set_xlabel('年齢(歳)',fontsize=12)
ax4.set_ylabel('人数',fontsize=12)

##目盛り
ax1.minorticks_on()
ax2.minorticks_on()
ax3.minorticks_on()
ax4.minorticks_on()

ax1.tick_params(which='both',direction='in')
ax2.tick_params(which='both',direction='in')
ax3.tick_params(which='both',direction='in')
ax4.tick_params(which='both',direction='in')

##グリッド
ax1.grid(which='major',axis='y',color='lightgray',linestyle='-')
ax2.grid(which='major',axis='y',color='lightgray',linestyle='-')
ax3.grid(which='major',axis='y',color='lightgray',linestyle='-')
ax4.grid(which='major',axis='y',color='lightgray',linestyle='-')

#タイトル
ax1.set_title('a', loc='left')
ax2.set_title('b', loc='left')
ax3.set_title('c', loc='left')
ax4.set_title('d', loc='left')

#描画
ax1.hist(age, bins=50, rwidth=0.8, color='lightblue')
ax2.hist(age, bins=30, rwidth=0.8, color='lightblue')
ax3.hist(age, bins=15, rwidth=0.8, color='lightblue')
ax4.hist(age, bins=10, rwidth=0.8, color='lightblue')

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](https://github.com/lvzeyu/chss_2025_assignment/blob/main/Assignment8/Figure2.png?raw=1)

```python
age_male = df[df['sex'] == 'male']['age']
age_female = df[df['sex'] == 'female']['age']

fig,ax = plt.subplots(1,2,figsize=(10,4))

ax2_1 = ax[0]
ax2_2 = ax[1]

#グラフの形式をそろえる
##軸ラベル
ax2_1.set_xlabel('乗客の年齢(歳)',fontsize=12)
ax2_1.set_ylabel('人数',fontsize=12)
ax2_2.set_xlabel('乗客の年齢(歳)',fontsize=12)
ax2_2.set_ylabel('人数',fontsize=12)

##目盛り
ax2_1.minorticks_on()
ax2_2.minorticks_on()

ax2_1.tick_params(which='both',direction='in')
ax2_2.tick_params(which='both',direction='in')

##グリッド
ax2_1.grid(which='major',axis='y',color='lightgray',linestyle='-')
ax2_2.grid(which='major',axis='y',color='lightgray',linestyle='-')


#タイトル
ax2_1.set_title('男性乗客')
ax2_2.set_title('女性乗客')

#描画
ax2_1.hist(age, bins=50,color='gray',label='すべての乗客')
ax2_1.hist(age_male,bins=50,color='blue',label='男性乗客')
ax2_1.legend(frameon =False)
ax2_2.hist(age, bins=50,color='gray',label='すべての乗客')
ax2_2.hist(age_female,bins=50,color='red',label='女性乗客')
ax2_2.legend(frameon=False)

plt.tight_layout()
plt.show()
```
