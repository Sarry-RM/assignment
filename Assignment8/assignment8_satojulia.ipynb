```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
```

## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

![](./Figure1.png)

```python
import matplotlib.pyplot as plt

# 年齢データ
age = df['age'].dropna()

# ヒストグラム
bins_list = [5, 10, 15, 20] 
titles = ['(a)', '(b)', '(c)', '(d)']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i in range(4):
    axes[i].hist(age, bins=range(0, int(age.max()) + bins_list[i], bins_list[i]), color='lightblue', edgecolor='black')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Count')
    axes[i].set_title(titles[i])
    axes[i].grid(axis='y')

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
import matplotlib.pyplot as plt

# NaNを除いた年齢・性別データ
df_age = df[['age', 'sex']].dropna()

# 各グループの年齢データ
all_age = df_age['age']
male_age = df_age[df_age['sex'] == 'male']['age']
female_age = df_age[df_age['sex'] == 'female']['age']

# binの設定を統一
bins = 20

# サブプロット作成（1行2列）
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# 男性：全体＋男性
axes[0].hist(all_age, bins=bins, color='gray', alpha=0.3, label='All', edgecolor='black')
axes[0].hist(male_age, bins=bins, color='blue', alpha=0.7, label='Male', edgecolor='black')
axes[0].set_title('Male Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
axes[0].legend()
axes[0].grid(axis='y')

# 女性：全体＋女性
axes[1].hist(all_age, bins=bins, color='gray', alpha=0.3, label='All', edgecolor='black')
axes[1].hist(female_age, bins=bins, color='red', alpha=0.7, label='Female', edgecolor='black')
axes[1].set_title('Female Age Distribution')
axes[1].set_xlabel('Age')
axes[1].legend()
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
