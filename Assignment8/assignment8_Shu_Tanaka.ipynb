## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

![](https://github.com/lvzeyu/chss_2025_assignment/blob/main/Assignment8/Figure1.png?raw=1)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")

ages = df['age'].dropna()

bin_widths = [2, 5, 10, 20]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, bw in enumerate(bin_widths):
  bins = range(int(ages.min()),int(ages.max()) + bw, bw)
  axes[i].hist(ages, bins=bins, color='lightblue', edgecolor='black')
  axes[i].set_title(f'Subplot ({chr(97+i)}) - Bin width: {bw}')
  axes[i].set_xlabel('年齢')
  axes[i].set_ylabel('人数')
  axes[i].grid(axis='y')

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](https://github.com/lvzeyu/chss_2025_assignment/blob/main/Assignment8/Figure2.png?raw=1)

```python
ages_all = df['age'].dropna()

ages_male = df[df['sex'] == 'male']['age'].dropna()
ages_female = df[df['sex'] == 'female']['age'].dropna()

bin_widths = 5
bins = range(int(ages_all.min()), int(ages_all.max()) + bin_widths, bin_widths)

plt.figure(figsize=(10, 6))

plt.hist(ages_all, bins=bins, color='gray', alpha=0.3, label='すべての乗客', edgecolor='black')
plt.hist(ages_male, bins=bins, color='blue', alpha=0.6, label='男性乗客', edgecolor='black')
plt.hist(ages_female, bins=bins, color='red', alpha=0.6, label='女性乗客', edgecolor='black')

plt.title('Age Distribution by Gender')
plt.xlabel('乗客の年齢（歳）')
plt.ylabel('人数')
plt.legend()
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
