```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
print(df.columns)

ages = df["age"].dropna()

bins_list = [5, 10, 20, 50]
labels = ['(a)', '(b)', '(c)', '(d)']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.hist(ages, bins=bins_list[i], color='lightblue', edgecolor='black')
    ax.set_title(f'{labels[i]} Histogram with {bins_list[i]} bins')
    ax.set_xlabel("Age")
    ax.set_ylabel("Frequency")
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")

ages_all = df["age"].dropna()
ages_female = df[df["sex"] == "female"]["age"].dropna()
ages_male = df[df["sex"] == "male"]["age"].dropna()


fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# 女性
axes[0].hist(ages_all, bins=30, color="lightgray", label="All", edgecolor='black')
axes[0].hist(ages_female, bins=30, color="pink", alpha=0.6, label="Female", edgecolor='black')
axes[0].set_title("Age Distribution: Female Passengers")
axes[0].set_ylabel("Frequency")
axes[0].legend()
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# 男性
axes[1].hist(ages_all, bins=30, color="lightgray", label="All", edgecolor='black')
axes[1].hist(ages_male, bins=30, color="lightblue", alpha=0.6, label="Male", edgecolor='black')
axes[1].set_title("Age Distribution: Male Passengers")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Frequency")
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
