## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

![](./Figure1.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'no-latex'])

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
age = df['age'].dropna()

bin_lists = [
    range(0, 81, 2),  
    range(0, 81, 3),   
    range(0, 81, 5),   
    range(0, 81, 10),  
]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()

labels = ['a', 'b', 'c', 'd']

for i, bins in enumerate(bin_lists):
    axs[i].hist(age, bins=bins, color='lightblue', alpha=0.7, rwidth=0.8, align='mid')
    axs[i].set_xlabel('Age(years)', fontsize=12)
    axs[i].set_ylabel('Counts', fontsize=12)
    axs[i].grid(axis='y', linestyle='--', alpha=0.7)
    axs[i].set_title(labels[i], loc='left', fontsize=14)

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'no-latex'])

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
df = df.dropna(subset=['age', 'sex'])  

age_all = df['age']
age_male = df[df['sex'] == 'male']['age']
age_female = df[df['sex'] == 'female']['age']

bins = range(0, 85, 2)

fig, axs = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

axs[0].hist(age_all, bins=bins, color='gray', label='All passengers')
axs[0].hist(age_male, bins=bins, color='blue', label='Male passengers')
axs[0].set_title("Male passengers", fontsize=14)
axs[0].set_xlabel("Passenger Age (years)", fontsize=12)
axs[0].set_ylabel("Count", fontsize=12)
axs[0].legend()
axs[0].grid(axis='y', linestyle='--', alpha=0.5)

axs[1].hist(age_all, bins=bins, color='gray', label='All passengers')
axs[1].hist(age_female, bins=bins, color='red', label='Female passengers')
axs[1].set_title("Female passengers", fontsize=14)
axs[1].set_xlabel("Passenger Age (years)", fontsize=12)
axs[1].legend()
axs[1].grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
