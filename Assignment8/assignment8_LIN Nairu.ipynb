## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

```python
!pip install matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm


plt.rcParams['font.family'] = 'MS Gothic'

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
age_data = df['age'].dropna()

fig, axes = plt.subplots(2, 2, figsize=(12, 6))
bin_widths = [2, 4, 6, 10]
labels = ['a', 'b', 'c', 'd']
axes = axes.flatten()

ymax_list = [90, 150, 220, 350]

for ax, bin_width, label, ymax in zip(axes, bin_widths, labels, ymax_list):
    bins = np.arange(0, 81 + bin_width, bin_width)
    ax.hist(age_data, bins=bins, color='lightblue', edgecolor='black',
            alpha=0.8, rwidth=0.9, histtype='bar') 
    ax.set_xlabel('年齢(歳)', fontsize=10)
    ax.set_ylabel('人数', fontsize=10)
    ax.set_title(label, loc='left', fontsize=12, fontweight='bold')
    ax.grid(axis='y')
    ax.set_ylim(0, ymax)  
    ax.tick_params(labelsize=8)

plt.tight_layout()
plt.show()

```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'MS Gothic'  

df = pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")
df = df.dropna(subset=['age'])  

age_all = df['age']
age_male = df[df['sex'] == 'male']['age']
age_female = df[df['sex'] == 'female']['age']

bins = np.arange(0, 81, 2)  

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

axes[0].hist(age_all, bins=bins, color='gray', label='すべての乗客')
axes[0].hist(age_male, bins=bins, color='blue', label='男性乗客')
axes[0].set_title('男性乗客')
axes[0].set_xlabel('乗客の年齢(歳)')
axes[0].set_ylabel('人数')
axes[0].legend()

axes[1].hist(age_all, bins=bins, color='gray', label='すべての乗客')
axes[1].hist(age_female, bins=bins, color='red', label='女性乗客')
axes[1].set_title('女性乗客')
axes[1].set_xlabel('乗客の年齢(歳)')
axes[1].legend()

for ax in axes:
    ax.grid(axis='y')

plt.tight_layout()
plt.show()
```
