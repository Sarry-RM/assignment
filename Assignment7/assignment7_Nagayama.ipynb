## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
abbrevs = pd.read_csv(r"C:\Users\haruz\chss_2025_assignment\Assignment7\Data\state-abbrevs.csv")
areas = pd.read_csv(r"C:\Users\haruz\chss_2025_assignment\Assignment7\Data\state-areas.csv")
population = pd.read_csv(r"C:\Users\haruz\chss_2025_assignment\Assignment7\Data\state-population.csv")
````

````python
abbrevs[:5]
````

````python
areas[:5]
````

````python
population[:5]
````

````python
df1 = pd.merge(abbrevs, areas, on = "state")
df1[:5]
````

````python
population2010 = population[population["year"] == 2010]
population2010
````

````python
total_population2010 = population2010[population2010["ages"] == "total"]
total_population2010[:5]
````

````python
df = pd.merge(df1, total_population2010, left_on = "abbreviation", right_on = "state/region")
df[:5]
````

````python
df["density"] = df["population"] / df["area (sq. mi)"]
df
````
