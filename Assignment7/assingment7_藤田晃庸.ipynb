```python
import pandas as pd

population = pd.read_csv("state-population.csv")
abbrevs = pd.read_csv("state-abbrevs.csv")
areas = pd.read_csv("state-areas.csv")
```

```python
import pandas as pd
pop_merged = pd.merge(population, abbrevs, how='left',
                      left_on='state/region', right_on='abbreviation')

pop_merged = pop_merged[pop_merged['state'].notnull()]
pop_merged = pop_merged.drop(columns='abbreviation')

pop_2010 = pop_merged[(pop_merged['year'] == 2010) & (pop_merged['ages'] == 'total')]

# 面積データ
merged = pd.merge(pop_2010, areas, how='left', on='state')
merged = merged[merged['area (sq. mi)'].notnull()]

# 人口密度の計算
merged['density'] = merged['population'] / merged['area (sq. mi)']

# .並べ替え
result = merged[['state', 'population', 'area (sq. mi)', 'density']]
result_sorted = result.sort_values(by='density', ascending=False)

print(result_sorted)
```
