## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
file_path2 = '/content/drive/MyDrive/データ/state-abbrevs.csv'
df2 = pd.read_csv(file_path2)
````

````python
file_path3 = '/content/drive/MyDrive/データ/state-areas.csv'
df3 = pd.read_csv(file_path3)
````

````python
file_path4 = '/content/drive/MyDrive/データ/state-population.csv'
df4 = pd.read_csv(file_path4)
````

````python
merged_df = pd.merge(df2, df3, on='state', how='outer')

final_df = pd.merge(merged_df, df4, left_on='abbreviation', right_on='state/region', how='outer')
````

````python
final_df
````

````python
#2010年アメリカの各州の人口密度
df_2010_state = final_df[(final_df['year'] == 2010) & (final_df['ages'] == 'total')]

df_2010_state['population_density'] = df_2010_state['population'] / df_2010_state['area (sq. mi)']

print(df_2010_state[['state', 'population_density']])
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df_grouped = final_df.groupby(['state/region', 'ages', 'year'])[['population', 'area (sq. mi)']].sum().reset_index()
df_grouped['population_density'] = df_grouped['population'] / df_grouped['area (sq. mi)']
df_grouped
````
